```python
import numpy as np
import matplotlib.pyplot as plt

# Data contoh: posisi benda (meter) vs waktu (detik)
t = np.array([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0])
x = np.array([0.0, 1.2, 4.8, 10.8, 19.2, 30.0, 43.2])

h = t[1] - t[0] # selang waktu = 0.5 detik
print("Selang h =", h)
```

Implementasi Tiga Metode

```python
n = len(t)
v_forward = np.zeros(n)
v_backward = np.zeros(n)
v_central = np.zeros(n)

# Forward difference
for i in range(n - 1):
    v_forward[i] = (x[i+1] - x[i]) / h

# Backward difference
for i in range(1, n):
    v_backward[i] = (x[i] - x[i-1]) / h

# Central difference
for i in range(1, n - 1):
    v_central[i] = (x[i+1] - x[i-1]) / (2 * h)

print("Kecepatan (central):", v_central)
```

Visualisasi Hasil

```python
plt.figure(figsize=(8, 4))
plt.plot(t[:-1], v_forward[:-1], 'r--o', label='Forward')
plt.plot(t[1:],  v_backward[1:], 'b--s', label='Backward')
plt.plot(t[1:-1],v_central[1:-1],'g-^',  label='Central')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Perbandingan Metode Diferensiasi Numerik')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Diferensiasi Data Berisik (Smoothing)

```python
from scipy.ndimage import uniform_filter1d

# Simulasi data berisik
np.random.seed(42)
t_n = np.linspace(0, 2*np.pi, 100)
y_clean = np.sin(t_n)
y_noisy = y_clean + 0.1 * np.random.randn(100)

# Smoothing sebelum diferensiasi
y_smooth = uniform_filter1d(y_noisy, size=5)
dy_noisy = np.gradient(y_noisy, t_n)
dy_smooth = np.gradient(y_smooth, t_n)
dy_exact = np.cos(t_n) # turunan analitik sin(x)

plt.plot(t_n, dy_noisy, 'r-', alpha=0.5, label='Tanpa smoothing')
plt.plot(t_n, dy_smooth, 'b-', label='Dengan smoothing')
plt.plot(t_n, dy_exact, 'g--',label='Eksak (cos x)')
plt.legend(); plt.grid(True); plt.show()
```
